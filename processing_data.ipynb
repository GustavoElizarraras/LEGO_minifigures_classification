```python
import os
import shutil
import pandas as pd
```

```python
df_metadata = pd.read_csv("data/metadata.csv")
df_metadata.head(5)
```

```python
figure_names = df_metadata[["class_id", "minifigure_name"]].set_index("class_id").to_dict()["minifigure_name"]
figure_names
```

```python
df_index = pd.read_csv("data/index.csv")
df_index.head(5)
```

```python
df_index["name"] = df_index["class_id"].map(figure_names)
df_index
```

```python
# Functions for processing the path
def change_name(path, new_name):
    separated = path.split("/")
    separated[1] = new_name
    return "/".join(separated[0:-1])

def only_dir(path):
    separated = path.split("/")
    return "/".join(separated[0:2])
```

```python
df_index["new_path"] = df_index.apply(lambda x: change_name(x["path"], x["name"]), axis=1)
df_index["only_dir"] = df_index.apply(lambda x: only_dir(x["path"]), axis=1)
change_path_name = dict(zip(df_index["only_dir"], df_index["new_path"]))
change_path_name["marvel/0007"] = "marvel/B. SPIDER-MAN"
change_path_name
```

```python
path = os.getcwd() + "/data/"
for root, dirs, files in os.walk(os.getcwd()):
    if path in root:
        lego_path = root.replace(path, "")
        if lego_path in change_path_name.keys():
            os.rename(path + lego_path, path + change_path_name[lego_path])
            shutil.move(path + change_path_name[lego_path], path)
```

```python
no_image_data = ["index.csv", "LICENSE", "metadata.csv", "test.csv", "archive.zip"]
for other_data in no_image_data:
    shutil.move(path + other_data, os.getcwd() + "/other_data")
shutil.move(path + "test", os.getcwd())
```
